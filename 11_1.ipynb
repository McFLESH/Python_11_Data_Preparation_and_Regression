```python
# Объявите функцию split_data(), которая принимает две переменные – название csv файла и название целевой переменной.
# Функция должна разделить датасет на тренировочные и тестовые данные в соотношении 3 к 1.
# На выходе пользователь получает результаты подсчётов в виде списка [features_train, features_test, target_train, target_test].


from sklearn.model_selection import train_test_split
import pandas as pd


def split_data(file, per):
    data = pd.read_csv(file)
    features = data.drop(per, axis=1)
    target = data[per]
    features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.25, random_state=42)
    return [features_train, features_test, target_train, target_test]
```
