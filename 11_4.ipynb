```python
# В этом задании вы будете работать с датасетом, который содержит данные о характеристиках автомобилей и их стоимости.
#Ваша задача - построить модель линейной регрессии для предсказания цен автомобилей (столбец MSRP), а затем сравнить её точность с  "глупой моделью".

#Необходимо реализовать всю логику программы в функции, которая в качестве аргумента принимает строку - имя файла.
#Загрузите данные из файла data.csv (сначала скачайте файл data.csv).
#Выберите в качестве предикторов следующие столбцы: Year, Engine HP, Engine Cylinders, highway MPG, city mpg и Popularity. Целевым признаком будет столбец MSRP.
#Обработайте пропущенные значения в признаках: замените их на средние значения соответствующих столбцов с помощью метода fillna.
#Разделите данные на тренировочную и тестовую выборки в пропорции 75/25, используя функцию train_test_split из библиотеки sklearn.
#Создайте модель линейной регрессии с помощью класса LinearRegression и обучите её на тренировочной выборке, вызвав метод fit.
#Выполните предсказания на тестовой выборке, используя метод predict.
#Вычислите метрику RMSE (среднеквадратичное отклонение) для оценки качества предсказаний модели. Для этого используйте функцию root_mean_squared_error из библиотеки sklearn.
#Реализуйте "глупую модель", которая для всех примеров предсказывает одно и то же значение — среднее значение целевого признака из тренировочной выборки.
#Вычислите RMSE для "глупой модели" аналогичным образом.
#Сравните RMSE линейной регрессии с RMSE "глупой модели". Выведите оба значения и разницу между ними.

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error
import pandas as pd


def engine(file):
    data = pd.read_csv(file)
    data = data.drop(['Make', 'Model', 'Engine Fuel Type', 'Transmission Type', 'Driven_Wheels', 'Number of Doors', 'Market Category', 'Vehicle Size', 'Vehicle Style'], axis=1)
    data = data.fillna(data.mean())

    features = data[['Year', 'Engine HP', 'Engine Cylinders', 'highway MPG', 'city mpg', 'Popularity']]
    target = data['MSRP']
    features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.25)

    model = LinearRegression()
    model.fit(features_train, target_train)
    result = model.predict(features_test)

    RMSE1 = round(root_mean_squared_error(target_test, result))

    dumb_model_result = [target_train.mean()] * len(target_test)
    RMSE2 = round(root_mean_squared_error(target_test, dumb_model_result))

    return f'RMSE линейной регрессии = {RMSE1}, RMSE "глупой модели" = {RMSE2}, разница = {RMSE2 - RMSE1}'
```
